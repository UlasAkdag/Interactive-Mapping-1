```python
import pandas as pd
import folium
```

### Read the data

```python
data = pd.read_excel("Most Visited Museums - Clean.xlsx")
del data["Unnamed: 0"]
```

```python
data
```

### Create the map

```python
m = folium.Map(tiles="OpenStreetMap", zoom_starts=2)
```

```python
m
```

### Add the locations to the map

```python
for index, row in data.iterrows():
    folium.Marker([row['Lat'], row['Lon']], popup="Name: "+row['Name']+
                  "<br> <br> No. of Yearly Visitors: "+row["Visitors Per Year"]).add_to(m)
```

```python
m
```

```python
m.save('museums_around_the_world.html')
```
